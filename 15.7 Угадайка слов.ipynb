–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ü–µ—Ä–µ–≤–æ–¥–∏—Ç–µ –≤—Å–µ —Å–∏–º–≤–æ–ª—ã –≤ –≤–µ—Ä—Ö–Ω–∏–π —Ä–µ–≥–∏—Å—Ç—Ä.

##### p.s. –î–æ–±–∞–≤–∏–ª –æ—Ç —Å–µ–±—è. –ï—Å–ª–∏ —é–∑–µ—Ä —É–≥–∞–¥–∞–ª –±—É–∫–≤—É –ø–æ–ø—ã—Ç–∫–∞ –Ω–µ –æ—Ç–Ω–∏–º–∞–µ—Ç—Å—è. 

```python
# —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ –≤–≤–æ–¥–∞ –±—É–∫–≤—ã –∏–ª–∏ —Å–ª–æ–≤–∞ –¥–ª—è –∏–≥—Ä—ã
def valid_answer_abc_and_word(guessed_words, guessed_letters):
    while True:
        user_input = input(f'–í–≤–µ–¥–∏—Ç–µ –±—É–∫–≤—É –∏–ª–∏ —Å–ª–æ–≤–æ —Ü–µ–ª–∏–∫–æ–º: ').upper()
        if (user_input.isalpha() and len(user_input) == len(word)):  # –µ—Å–ª–∏ –≤–≤–µ–¥–µ–Ω–æ —Å–ª–æ–≤–æ –∏ –æ–Ω–æ –¥–ª–∏–Ω–Ω–æ–π —Å –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–µ 
            if user_input in guessed_words:  # –µ—Å–ª–∏ —É–∂–µ —Ç–∞–∫–æ–µ —Å–ª–æ–≤–æ –ø—Ä–æ–±–æ–≤–∞–ª
                print("–¢—ã —É–∂–µ –≤–≤–æ–¥–∏–ª —ç—Ç–æ —Å–ª–æ–≤–æ, –ø–æ–ø—Ä–æ–±—É–π —Å–Ω–æ–≤–∞")
            else:
                return(user_input)
        elif (user_input.isalpha() and len(user_input) == 1):  # –µ—Å–ª–∏ –≤–≤–µ–¥–µ–Ω–∞ –±—É–∫–≤–∞ 
            # print("USER_INPUT_OK", guessed_letters)  # –ø—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞–±–æ—Ç—ã
            if user_input in guessed_letters:  # –µ—Å–ª–∏ —É–∂–µ –≤–≤–æ–¥–∏–ª —Ç–∞–∫—É—é –±—É–∫–≤—É
                print("–¢—ã —É–∂–µ –≤–≤–æ–¥–∏–ª —ç—Ç—É –±—É–∫–≤—É, –ø–æ–ø—Ä–æ–±—É–π —Å–Ω–æ–≤–∞")
            else:
                return(user_input)
        else:
            print(f'–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –æ—Ç–≤–µ—Ç, –≤–≤–µ–¥–∏—Ç–µ –±—É–∫–≤—É –∏–ª–∏ —Å–ª–æ–≤–æ —Ü–µ–ª–∏–∫–æ–º. –°–ª–æ–≤–æ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ {len(word)} –±—É–∫–≤')
```

```python
def play(word):
    word_completion = ['_']* len(word) # —Å–ø–∏—Å–æ–∫, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π —Å–∏–º–≤–æ–ª—ã _ –Ω–∞ –∫–∞–∂–¥—É—é –±—É–∫–≤—É –∑–∞–¥—É–º–∞–Ω–Ω–æ–≥–æ —Å–ª–æ–≤–∞
    tries = 6                          # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ø—ã—Ç–æ–∫
    guessed = False                    # —Å–∏–≥–Ω–∞–ª—å–Ω–∞—è –º–µ—Ç–∫–∞
    guessed_letters = []               # —Å–ø–∏—Å–æ–∫ —É–∂–µ –Ω–∞–∑–≤–∞–Ω–Ω—ã—Ö –±—É–∫–≤
    guessed_words = []                 # —Å–ø–∏—Å–æ–∫ —É–∂–µ –Ω–∞–∑–≤–∞–Ω–Ω—ã—Ö —Å–ª–æ–≤                  
    print('–î–∞–≤–∞–π –∏–≥—Ä–∞—Ç—å –≤ —É–≥–∞–¥–∞–π–∫—É —Å–ª–æ–≤!')
    print(display_hangman(tries))
    # —Ç–µ–ª–æ –∏–≥—Ä—ã
    while True:
        print (f'–ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —Å–ª–æ–≤–æ {"".join(word_completion)} —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ {len(word)} –±—É–∫–≤.', " \n " 
                f'–£ —Ç–µ–±—è –æ—Å—Ç–∞–ª–æ—Å—å {tries} –ø–æ–ø—ã—Ç–æ–∫ –æ—Ç–≥–∞–¥–∞—Ç—å —Å–ª–æ–≤–æ.')
        user_input = valid_answer_abc_and_word(guessed_words, guessed_letters)
        if len(user_input) == len(word):  # User –Ω–∞–∑–≤–∞–ª —Å–ª–æ–≤–æ —Ü–µ–ª–∏–∫–æ–º
            if user_input == word:
                guessed = True  # —Ñ–ª–∞–≥ —É–≥–∞–¥–∞–ª –∏ –≤—ã—Ö–æ–¥–∏–º –∏–∑ —Ç–µ–ª–∞ –∏–≥—Ä—ã
                break
            else:
                guessed_words.append(user_input)
                tries -= 1
        else:
            if user_input in word:  # –±—É–∫–≤–∞ —Å–æ–≤–ø–∞–ª–∞
                guessed_letters.append(user_input)  # –¥–æ–±–∞–≤–ª—è–µ–º –≤ —Å–ø–∏—Å–æ–∫ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –±—É–∫–≤
                # print(guessed_letters)  # –æ—Ç–ª–∞–¥–æ—á–Ω—ã–π –≤—ã–≤–æ–¥
                letter_word = 0  # —Ç–æ—á–∫–∞ —Å—Ç–∞—Ä—Ç–∞ –ø–æ–∏—Å–∫–∞ –±—É–∫–≤—ã
                while True:
                    letter_word = word.find(user_input, letter_word)  # –ø–æ–∏—Å–∫ –ø–æ–∑–∏—Ü–∏–∏ –±—É–∫–≤—ã —Å–æ —Å—Ç–∞—Ä—Ç–æ–≤–æ–π —Ç–æ—á–∫–æ–π letter_word
                    # print(letter_word ,word_completion , word )
                    if letter_word == -1:  # –µ—Å–ª–∏ –≤—Ö–æ–∂–¥–µ–Ω–∏–π –±–æ–ª—å—à–µ –Ω–µ–Ω–∞–π–¥–µ–Ω–æ 
                        break
                    else:  # –µ—Å–ª–∏ –Ω–∞—à–ª–æ—Å—å —Ç–æ –æ—Ç–∫—Ä—ã–≤–∞–µ–º —ç—Ç—É –±—É–∫–≤—É –≤ –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–º —Å–ª–æ–≤–µ
                        # print(letter_word, word_completion[letter_word])  # –æ—Ç–ª–∞–¥–æ—á–Ω—ã–π –≤—ã–≤–æ–¥
                        word_completion[letter_word] = user_input  # –º–µ–Ω—è–µ–º "_" –Ω–∞ –±—É–∫–≤—É
                        letter_word += 1  # —Å–¥–≤–∏–≥–∞–µ–º —Ç–æ—á–∫—É —Å—Ç–∞—Ä—Ç–∞ –Ω–∞ + 1, —á—Ç–æ–±—ã –Ω–µ –∑–∞—Ü–∏–∫–ª–∏–ª–∞—Å—å –Ω–∞ –Ω–∞–π–¥–µ–Ω–Ω–æ–π
            else:
                guessed_letters.append(user_input) # –µ—Å–ª–∏ –Ω–µ —É–≥–∞–¥–∞–ª –¥–æ–±–∞–≤–ª—è–µ–º –≤ —Å–ø–∏—Å–æ–∫ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –±—É–∫–≤
                tries -= 1

        if "".join(word_completion) == word:  # —É–≥–∞–¥–∞–ª –≤—Å–µ –±—É–∫–≤—ã
            guessed = True
            break
        print(display_hangman(tries))  # –≤—ã–≤–æ–¥–∏–º —Ç–µ–∫—É—â–µ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ –∏–≥—Ä—ã
        if tries == 0:  # –ù–∞ —Å–ª—É—á–∞–π GAME OVER'–∞
            print(f"GAME OVER \U0001F635 –ë—ã–ª–æ –∑–∞–≥–∞–¥–∞–Ω–æ —Å–ª–æ–≤–æ {word}.")
            break
    if guessed == True:
            print('–ü–æ–∑–¥—Ä–∞–≤–ª—è–µ–º, –≤—ã —É–≥–∞–¥–∞–ª–∏ —Å–ª–æ–≤–æ! –í—ã –ø–æ–±–µ–¥–∏–ª–∏!')
```
