```python
def play(word):
    word_completion = ['_']* len(word) # список, содержащий символы _ на каждую букву задуманного слова
    tries = 6                          # количество попыток
    guessed = False                    # сигнальная метка
    guessed_letters = []               # список уже названных букв
    guessed_words = []                 # список уже названных слов                  
    print('Давай играть в угадайку слов!')
    print(display_hangman(tries))
    # тело игры
    while True:
        print (f'Загаданное слово {"".join(word_completion)} состоит из {len(word)} букв.', " \n " 
                f'У тебя осталось {tries} попыток отгадать слово.')
        user_input = valid_answer_abc_and_word(guessed_words, guessed_letters)
        if len(user_input) == len(word):  # User назвал слово целиком
            if user_input == word:
                guessed = True  # флаг угадал и выходим из тела игры
                break
            else:
                guessed_words.append(user_input)
                tries -= 1
        else:
            if user_input in word:  # буква совпала
                guessed_letters.append(user_input)  # добавляем в список введенных букв
                # print(guessed_letters)  # отладочный вывод
                letter_word = 0  # точка старта поиска буквы
                while True:
                    letter_word = word.find(user_input, letter_word)  # поиск позиции буквы со стартовой точкой letter_word
                    # print(letter_word ,word_completion , word )
                    if letter_word == -1:  # если вхождений больше ненайдено 
                        break
                    else:  # если нашлось то открываем эту букву в загаданном слове
                        # print(letter_word, word_completion[letter_word])  # отладочный вывод
                        word_completion[letter_word] = user_input  # меняем "_" на букву
                        letter_word += 1  # сдвигаем точку старта на + 1, чтобы не зациклилась на найденной
            else:
                guessed_letters.append(user_input) # если не угадал добавляем в список введенных букв
                tries -= 1

        if "".join(word_completion) == word:  # угадал все буквы
            guessed = True
            break
        print(display_hangman(tries))  # выводим текущее состояние игры
        if tries == 0:  # На случай GAME OVER'а
            print(f"GAME OVER \U0001F635 Было загадано слово {word}.")
            break
    if guessed == True:
            print('Поздравляем, вы угадали слово! Вы победили!')
```
