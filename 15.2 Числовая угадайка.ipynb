### Угадайка чисел
Описание проекта: программа генерирует случайное число в диапазоне от 1 до 100 и просит пользователя угадать это число. Если догадка меньше случайного числа, то программа должна вывести сообщение 'Слишком мало, попробуйте еще раз'. Если пользователь угадывает число, то программа должна поздравить его и вывести сообщение 'Вы угадали, поздравляем!'.

```python
number = random.randint(1, 100)
# print(number)
x = int(input("Введите число: "))
while x != number:
    if x > number:
        print("Слишком много, попробуйте еще раз")
        x = int(input("Введите число: "))
    elif x < number:
        print("Слишком мало, попробуйте еще раз")
        x = int(input("Введите число: "))
    if x == number:
        print("Вы угадали, поздравляем!")    
```

### Оптимальная стратегия угадывания числа
Чтобы гарантированно угадать задуманное число от 1 до 100 потребуется не более 7 попыток.<br>

Оптимальный алгоритм угадывания: положим left = 1 и right = 100.<br>

Называем число, равное middle = (left + right) // 2;<br>
Если число middle равно задуманному числу, то мы угадали!;<br>
Если число middle меньше задуманного числа, то положим left = middle + 1 и продолжим алгоритм;<br>
Если число middle больше задуманного числа, то положим right = middle - 1 и продолжим алгоритм.<br>
Поскольку на каждой итерации мы отбрасываем половину чисел, то гарантировано угадаем задуманное число за величину, равную 
log n (двоичный логарифм) округленную до целого в большую сторону. При n=100 получаем 7 попыток.

```python
from random import randint

left, right = 1, 100  # диапазон от генерации
middle = (left + right) // 2  # начальное число берем как середина предполагаемого диапазона 
count = 1
num = randint(1, 100)  # генерируем случайное число
while middle != num:  # цикл пока не найдем число
    if middle > num:  # если предполагаемое число меньше загаданного
        right = middle - 1  # берем правую часть как половину минус 1 
        middle = (left + right) // 2  # получаем новое число предсказания  
        count += 1  # счетчик поиска + 1
    elif middle < num:  # если предполагаемое число больше загаданного 
        left = middle + 1  # берем левую часть как половину плюс 1
        middle = (left + right) // 2  # получаем новое число предсказания  
        count += 1  # счетчик поиска + 1
print(f'Количество попыток: {count}, это число: {middle}')
print(f'Было загадано число: {num}')
```
